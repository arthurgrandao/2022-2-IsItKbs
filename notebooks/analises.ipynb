```python
import pandas as pd
```

```python
def letter_counter(string, type=None):
    if (type == 'c'):
        return {letter: str(string).count(letter) for letter in 'bcdfghjklmnpqrstvwxyz'}
    elif (type == 'v'):
        return {letter: str(string).count(letter) for letter in 'aeiou'}  
    return {letter: str(string).count(letter) for letter in 'abcdefghijklmnopqrstuvwxyz'}

def type_counter(string, type):    
    return sum(letter_counter(string, type).values())

def type_ratio(string, type):
    return type_counter(string, type)/len(string)

def repetitions_character(repetitions, caractere):
        x = repetitions.get(caractere)
        return x if x != None else 0
    
def repetitions(string):
    try:
        resultados = []
        size = len(string)
        valor = 1

        for l in range(0, size):
            if l <= size-2 and string[l] == string[l+1]:
                valor += 1
            else:
                resultados.append(tuple((string[l], valor)))
                valor = 1
        return dict(sorted(resultados))
    except:
        return {}

def repetitions_type(string, type):
    if type == 'c':
        letters = 'bcdfghjklmnpqrstvwxyz'
    else:
        letters = 'aeiou'
    
    resultados = []
    size = len(string)
    valor = 0

    for l in string:
        if l in letters:
            valor += 1
        else:
            resultados.append(valor)
            valor = 0
    resultados.append(valor)
    
    return max(resultados)

def bigram_counter(lista):
  dic = {}
  for i in lista:
    if i in dic:
      dic[i] +=1
    else:
      dic[i] =1
  return dic

def bigrams(string):
    bigrams = []
    for i in range(len(string)-1):
        bigrams.append(string[i]+string[i+1])

    return bigram_counter(bigrams)

def bigram_max_occurance(string):
    try:
        return (sorted(bigrams(string).values(), reverse=True))[0]
    except:
        return 0
```

# Análise de média de vogais e consoantes

```python
aux = df.copy()
aux["len"] = aux["words"].apply(lambda x : len(str(x)))
aux["vowel rt"] = aux["words"].apply(lambda x : type_ratio(str(x), 'v'))
aux["consonant rt"] = aux["vowel rt"].apply(lambda x : 1-x)
```

```python
aux.sort_values(by=['len', 'words'], inplace=True, ignore_index=True)
aux
```

```python
newdf = pd.DataFrame()
rows = ["min vowel rt", "max vowel rt", "min consonant rt", "max consonant rt"]
newdf.insert(0, "len", rows)
```

```python
def analysys_by_len_rt(df, newdf, len, column):
    to_append = []
    index = keys.index(f'{len}')
    
    try:    
        # vrt_min
        to_append.append(min(df['vowel rt'][items[index][1]:items[index+1][1]]))
        # vrt_max
        to_append.append(max(df['vowel rt'][items[index][1]:items[index+1][1]]))
        # crt_min
        to_append.append(min(df['consonant rt'][items[index][1]:items[index+1][1]]))
        # crt_max
        to_append.append(max(df['consonant rt'][items[index][1]:items[index+1][1]]))

    except:
        # vrt_min
        to_append.append(df['vowel rt'][items[index][1]])
        # vrt_max
        to_append.append(df['vowel rt'][items[index][1]])
        # crt_min
        to_append.append(df['consonant rt'][items[index][1]])
        # crt_max
        to_append.append(df['consonant rt'][items[index][1]])
    
    newdf.insert(int(column), f"{len}", to_append)
```

```python
column = 0
for i in keys:
    analysys_by_len_rt(aux, newdf, i, column)
    column += 1

newdf.set_index('len', inplace=True)

newdf
```

```python
newdf.to_csv("../data/processed/analise_v&c_rt.csv")
```

# Análise cadeia de repetições

```python
aux = df.copy()
aux["len"] = aux["words"].apply(lambda x : len(str(x)))
for letter in letters:
    aux[f"{letter}"] = aux["words"].apply(lambda x : repetitions_character(repetitions(str(x)), letter))
```

```python
aux.sort_values(by=['len', 'words'], inplace=True, ignore_index=True)
aux
```

```python
newdf = pd.DataFrame()
newdf.insert(0, "len", letters)
```

```python
def analysys_by_len_repetition(df, newdf, len, column):
    to_append = []
    index = keys.index(f'{len}')

    try:    
        for letter in letters:
            to_append.append(max(df[f'{letter}'][items[index][1]:items[index+1][1]]))
    except:
        for letter in letters:
            to_append.append(df[f'{letter}'][items[index][1]])
    
    newdf.insert(int(column), f"{len}", to_append)
```

```python
column = 0
for i in keys:
    analysys_by_len_repetition(aux, newdf, i, column)
    column += 1

newdf.set_index('len', inplace=True)
newdf
```

```python
newdf.to_csv("../data/processed/analise_len_cadeias.csv")
```

# Análise cadeia de repetições de consoantes e vogais

```python
aux = df.copy()
aux["len"] = aux["words"].apply(lambda x : len(str(x)))
aux["vowel rp"] = aux["words"].apply(lambda x : repetitions_type(str(x), 'v'))
aux["consonant rp"] = aux["words"].apply(lambda x : repetitions_type(str(x), 'c'))
```

```python
newdf = pd.DataFrame()
rows = ["vowel rp", "consonant rp"]
newdf.insert(0, "len", rows)
```

```python
def analysys_by_len_repetition_type(df, newdf, len, column):
    to_append = []
    index = keys.index(f'{len}')
    
    try:    
        to_append.append(max(df['vowel rp'][items[index][1]:items[index+1][1]]))
        to_append.append(max(df['consonant rp'][items[index][1]:items[index+1][1]]))
        
    except:
        to_append.append(df['vowel rp'][items[index][1]])
        to_append.append(df['consonant rp'][items[index][1]])
    
    newdf.insert(int(column), f"{len}", to_append)
```
